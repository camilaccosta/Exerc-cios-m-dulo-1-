O arquivo googleplaystore.csv é um dataset que contém diversas informações sobre aplicativos disponíveis para download na Google Play Store. Logo após, verifique se existem dados duplicados e/ou ausentes nesse dataset.

Caso existam dados duplicados, remova-os. Caso existam dados ausentes, você pode removê-los ou pensar em alguma maneira de preenchê-los.



```python
import pandas as pd 
```

```python
google_play = pd.read_csv("googleplaystore.csv")
```

**Verifique se existem dados duplicados e/ou ausentes nesse dataset.**

**Caso existam dados duplicados, remova-os.**

```python
google_play.drop_duplicates("App", inplace=True)
```

**Caso existam dados ausentes, você pode removê-los ou pensar em alguma maneira de preenchê-los.**

```python
google_play.dropna(inplace=True)
```

O arquivo houses_to_rent.csv consiste em um dataset que contém dados sobre casas para alugar no Brasil. Carregue-o em um objeto DataFrame e verifique os dados contidos nas colunas desse DF (você pode utilizar os métodos head, tail e info). Logo após, verifique se existem dados duplicados e/ou ausentes nesse dataset.

Caso existam dados duplicados, remova-os. Caso existam dados ausentes, você pode removê-los ou pensar em alguma maneira de preenchê-los.

```python
# Realizando a leitura do arquivo 
houses_to_rent=pd.read_csv("houses_to_rent.csv" ) 
```

```python
# Fazendo um filtro a fim de verificar se existem linhas duplicadas.

houses_to_rent[houses_to_rent.duplicated()== True]
```

*No output anterior foi possível verificar que 358 linhas eram duplicadas.* 

**Caso existam dados duplicados, remova-os.**

```python
#Removendo as linhas duplicadas. O parâmetro keep = "first", faz com que a primeira linha duplicada seja mantida e as demais apagadas. 
houses_to_rent.drop_duplicates(keep="first")
```

```python
#Verificando se tem algum valor NaN no Data Frame.
houses_to_rent.isnull().values.any()
```
